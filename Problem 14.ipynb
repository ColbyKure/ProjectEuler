# Project Euler Problem 14

<b>The following iterative sequence is defined for the set of positive integers:

    n → n/2      (if n is even)
    n → 3n + 1   (if n is odd)

Using the rule above and starting with 13, we generate the following sequence:

13 → 40 → 20 → 10 → 5 → 16 → 8 → 4 → 2 → 1
It can be seen that this sequence (starting at 13 and finishing at 1) contains 10 terms. Although it has not been proved yet (Collatz Problem), it is thought that all starting numbers finish at 1.

Which starting number, under one million, produces the longest chain?

NOTE: Once the chain starts the terms are allowed to go above one million.</b>

```python
def get_next(k):
    if (k % 2) == 0:
        return int(k/2)
    return int( (3 * k) + 1 )


import time
start_time = time.time()



max = 1000000
max_len = 0
#max_seq = []
seq_len = 0
#seq = []
for i in range(max, 2, -1):
    if (i < int(max/2)) & ((i % 2) == 0):
        continue
    if (3 * i + 1) < max:
        continue
    #seq = [i]
    seq_len = 1
    curr = i
    while True:
        curr = get_next(curr)
        #seq.append(curr)
        seq_len += 1
        if curr == 1:
            break
    if max_len <= seq_len:
        max_len = seq_len
        #max_seq = seq
        print("with starting point ", i, " found sequence of ", seq_len)

print(max_len)        
        
        
        
print("--- %s seconds ---" % (time.time() - start_time))
```
